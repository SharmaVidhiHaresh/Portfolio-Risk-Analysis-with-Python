Although this can be achieved using numy and limitation of list is overcomed,numpy is quite tedious. Thus, you can use pandas library and construct a data frame. Dictionary values would be inserted in the data frame.

```python
import pandas as pd
prices = pd.DataFrame({"Blue" : [8.70, 8.91,8.71,8.43,8.73]                        ,
                       "Orange" :[1.66,1.08,1.71,1.59,1.11]})
```

```python
prices
```

```python
prices.iloc[1:]
```

```python
prices.iloc[1:]/prices.iloc[:-1]
```

Now, this would lead to an alignment issue. Thus, you would have to include .value after iloc[1:].values

```python
prices.iloc[1:].values/prices.iloc[:-1]
```

Another method to do this is by including shift(position to remove)

```python
prices.shift(1)
```

```python
prices/prices.shift(1)-1
```
